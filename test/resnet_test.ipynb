```python
import torchvision.models as models
from torchsummary import summary
```

```python
resnet18 = models.resnet18(pretrained=True)
summary(resnet18, (3, 320, 180), device="cpu")
```

```python
import torch
out = resnet18(torch.zeros(1, 3, 32, 180))
print(out.shape)
```

```python
newmodel = torch.nn.Sequential(*(list(resnet18.children())[:]))
summary(newmodel, (3, 320, 180))
out = newmodel(torch.zeros(1, 3, 180, 320))
print(out.shape)
```
